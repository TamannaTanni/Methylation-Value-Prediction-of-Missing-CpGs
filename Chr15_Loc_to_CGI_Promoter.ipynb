```
import pandas as pd
import pyranges as pr
```

```
chr15_cgi = pd.read_csv('Loc_to_CGI.csv')
chr15_cgi
```

```
promoter = pr.read_bed("hg38v79_genes_TSS2kb.bed", as_df=True)
promoter = promoter.loc[promoter['Chromosome'] == 'chr15']
promoter.to_csv('chr15_promoter.csv', index = False)
chr15_promoter = pd.read_csv('chr15_promoter.csv')
chr15_promoter
```

```
chr15_cgi['Loc_to_promoter'] = ""
for i in range(0, len(chr15_promoter)):
  chr15_cgi.loc[(chr15_cgi['start position'] >= chr15_promoter.iloc[i][1]) & (chr15_cgi['end position'] <= chr15_promoter.iloc[i][2]), 'Loc_to_promoter'] = 1
chr15_cgi.loc[chr15_cgi['Loc_to_promoter'] == '' , 'Loc_to_promoter'] = 0
chr15_promoter1 = chr15_cgi
chr15_promoter1.to_csv('Chr15_Loc_to_CGI_P.csv')
Chr15_Loc_to_CGI_P = pd.read_csv('Chr15_Loc_to_CGI_P.csv')
Chr15_Loc_to_CGI_P
```

```
Chr15_Loc_to_CGI_P1 = Chr15_Loc_to_CGI_P.loc[Chr15_Loc_to_CGI_P['Loc_to_promoter'] == 1]
Chr15_Loc_to_CGI_P1 = Chr15_Loc_to_CGI_P1.reset_index()
Chr15_Loc_to_CGI_P1
```
